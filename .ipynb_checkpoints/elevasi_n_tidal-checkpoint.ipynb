# **Koreksi Elevasi & Tidal dalam Metode Gaya Berat**
***


Script ini dibuat untuk menghitung koreksi elevasi dan tidal yang didapat dari dua alat ukur elevasi yang berbeda, yaitu Altimeter dan GPS. Terdapat 2 kelompok yang melakukan akuisisi data gravitasi. User dapat memilih kelompok 1 atau 2 yang akan diterapkan koreksi elevasi dan tidak terhadapnya.
***

###  Import fungsi/modul yang diperlukan

```python
import pandas as pd
from pyproj import Proj
import numpy as np
from datetime import datetime
from tidegravity import solve_point_corr, solve_tide_df
from matplotlib import pyplot as plt
from datetime import timedelta

# visualisasi template
plt.style.use('ggplot')
plt.rcParams["font.size"] = "15"
```

### Definisi variabel untuk membantu pengolahan data

```python
# elevasi base sebagai acuan
z_acuan = 766

# Kelompok akuisisi data
kelompok = 1

# Alat pengukur elevasi, opsinya 'Alti' dan 'GPS' untuk script ini
alat = 'Alti'

# penyederhanaan variabel bergantung alat pengukur elevasi
if alat == 'Alti':
    idx = 3
    header = 'HH Altimeter (m)'
elif alat == 'GPS':
    idx = 4
    header = 'HH GPS (m)'
```

# 1. LOAD DATA EXCEL HH GPS dan ALTIMETER

## 1.1 Data HH GPS dan Altimeter setiap stasiun

```python
# file ecel data stasiun untuk kelompok tertentu, di dalamnya ada worksheet GPS_Alti_sta dan GPS_Alti_base
df = pd.ExcelFile(f"dataset\\raw data {kelompok}.xlsx")

# membaca file excel dalam worksheet stasiun
df_stasiun = df.parse('GPS_Alti_sta')

# menampilkan 5 baris pertama tabel
df_stasiun.head()
```

## 1.2 Data HH GPS dan Altimeter di base

```python
# membaca file excel dalam worksheet HH_GPS_Alti
df_base = df.parse('GPS_Alti_base')

# menampilkan 5 baris pertama dalam tabel
df_base.head()
```

# 2. KOREKSI DATA ELEVASI

Deskripsi variabel yang digunakan dalam koreksi data elevasi:
> z_base : nilai elevasi di base pada waktu akuisisi data tertentu

> z_sta_base : nilai perbedaan elevasi antara stasiun dan base di waktu yang sama

> z_relatif : nilai elevasi relatif z_sta_base terhadap z_sta_base pertama

> z_drift : nilai koreksi drift untuk z_relatif

> z_terkoreksi : nilai elevasi z_relatif yang ditambah oleh z_drift

> z_true : elevasi sebenarnya, diambil berdasarkan nilai elevasi acuan


untuk interpolasi linier, digunakan beberapa variabel:
> z_akhir : nilai elevasi batas atas yang digunakan pada saat interpolasi linier

> z_awal : nilai elevasi batas bawah yang digunakan pada saat interpolasi linier

> t_akhir : waktu batas atas

> t_awal : waktu batas bawah


## 2.1 menghitung z_base

```python
# inisiasi variabel untuk z_base di waktu akuisisi data
z_base = []

# iterasi untuk mendapatkan nilai z_base di waktu akuisisi data setiap stasiun
for i in range(1, len(df_stasiun) - 1):
    # waktu pengukuran di stasiun
    time = df_stasiun['Time'][i]
    
    # waktu pengukuran di base yang terdekat ke waktu pengukuran di stasiun (batas bawah)
    time_sebelum = df_base[df_base['Time'] <= time]
    time_sebelum = time_sebelum.loc[[time_sebelum.index[-1]]].values

    # waktu pengukuran di base yang terdekat ke waktu pengukuran di stasiun (batas atas)
    time_sesudah = df_base[df_base['Time'] > time]
    time_sesudah = time_sesudah.loc[[time_sesudah.index[0]]].values

    # batas atas dan bawah elevasi di base
    z_akhir = time_sesudah[0, idx]
    z_awal =  time_sebelum[0, idx]

    # batas atas dan bawah waktu pengukuran di base
    t_akhir = time_sesudah[0, 2]
    t_awal = time_sebelum[0, 2]

    # melakukan interpolasi linier untuk mendapatkan nilai z_base pada waktu akuisisi data setiap stasiun
    z_base.append((z_akhir - z_awal) / ((t_akhir - t_awal) / timedelta(minutes=1)) * ((time - t_awal) / timedelta(minutes=1)) + z_awal)
```

**kondisi di bawah ini dilakukan untuk mengatasi masalah pencatatan elevasi pertama di stasiun dimulai lebih awal dari pencatatan elevasi pertama di base serta pencatatan terakhir di stasiun setelah pencatatan elevasi terakhir di base.**

```python
# nilai z_base pada pengukuran stasiun pertama disamakan dengan data elevasi acuan
z_base.insert(0, z_acuan)

# nilai z_base pada pengukuran stasiun terakhir disamakan dengan data elevasi acuan
z_base.append(z_acuan)
```

<!-- ## 2.2 menghitung z_true -->

```python
# nilai perbedaan elevasi antara stasiun dan base di waktu yang sama
z_sta_base = df_stasiun[header] - z_base

# nilai elevasi relatif z_sta_base terhadap z_sta_base pertama
z_relatif = z_sta_base - z_sta_base[0]

# waktu pengukuran di base yang terdekat ke waktu pengukuran di stasiun (batas bawah)
time_sebelum = df_base[df_base['Time'] <= time]
time_sebelum = time_sebelum.loc[[time_sebelum.index[-1]]].values

# waktu pengukuran di base yang terdekat ke waktu pengukuran di stasiun (batas atas)
time_sesudah = df_base[df_base['Time'] > time]
time_sesudah = time_sesudah.loc[[time_sesudah.index[0]]].values

# batas atas dan bawah elevasi relatif
z_akhir = z_relatif[z_relatif.index[-1]]
z_awal =  z_relatif[z_relatif.index[0]]

# batas atas dan bawah waktu pengukuran di stasiun
t_akhir = df_stasiun['Time'][df_stasiun.index[-1]]
t_awal = df_stasiun['Time'][df_stasiun.index[0]]

# menghitung koreksi drift untuk elevasi
z_drift = []
for i in range(len(df_stasiun)):
    z_drift.append((z_akhir - z_awal) / ((t_akhir - t_awal) / timedelta(minutes=1)) * ((df_stasiun['Time'][df_stasiun.index[i]] - t_awal) / timedelta(minutes=1)))

# menghitung elevasi relatif yang dikoreksi menggunakan koreksi drift
z_terkoreksi = z_relatif - z_drift

# menghitung elevasi sebenarnya
z_true = z_terkoreksi + (z_acuan - z_terkoreksi[0])
```

## 2.3 Tabel lengkap koreksi elevasi

```python
# membuat tabel elevasi
df_elevasi = pd.DataFrame({
#     'ID STS': df_stasiun['ID STS'],
    f'{header}': df_stasiun[header],
    'z base (m)': z_base,
    'z sta - z base (m)': z_sta_base,
    'z relatif (m)': z_relatif,
    'z drift (m)': z_drift,
    'z terkoreksi (m)': z_terkoreksi,
    'z true (m)': z_true
})

# mengganti indeks menjadi nama stasiun
df_elevasi.index = df_stasiun['ID STS']

# menampiklan 5 baris pertama dalam tabel
df_elevasi
```

## 2.3 plot koreksi elevasi

```python
plt.figure(figsize=(10,8))
df_base.index = df_base['Time']
df_base.iloc[:,idx].plot()

plt.title(f"{header[0:-4]} Base")
plt.ylabel("Elevasi (m)")
plt.show()
```

# 3. KOREKSI TIDAL

Perhitungan koreksi tidal mengunakan python package tidegravity. Beberapa inputnya adalah koordinat dalam lat/lon, altitude, dan juga waktu pengukuran. Catatan penting adalah waktu yang digunakan harus dalam UTC+0. Oleh karenanya, waktu pengukuran di stasiun kali ini dikurangi 7 jam.

Perhitungan koreksi tidal ini merupakan implementasi dari `Longman Earth Tide Calculator`. Beberapa fungsi yang ada di dalamnya adalah:
> **solve_tide_df()** : untuk menghitung nilai koreksi tidal di setiap titik di waktu tertentu. fungsi ini digunakan untuk menentukan nilai koreksi tidal di setiap stasiun pada saat pengukuran

> **solve_point_corr()** : menghitung nilai koreksi tidal di 1 titik dalam rentang waktu tertentu. fungsi ini digunakan untuk menentukan variasi koreksi tidal di base


## 3.1 Proyeksi sistem koordinat UTM ke Lat/Lon setiap stasiun

```python
# proyeksi koordinat
myProj = Proj(proj='utm', ellps='WGS84', zone='48S', south=True)
lon, lat = myProj(df_stasiun['UTM X'].values, df_stasiun['UTM Y'].values, inverse=True)

# membuat tabel hasil proyeksi koordinat dan altitude dari nilai koreksi elevasi
df_lonlat = pd.DataFrame(np.c_[lon, lat], columns=['lon', 'lat'])
df_lonlat = df_lonlat.set_index(df_stasiun['Time'] - timedelta(hours=7))
df_lonlat['alt'] = df_elevasi['z true (m)'].values

# menampilkan 5 baris pertama dalam tabel
df_lonlat.head()
```

## 3.2 Menghitung koreksi tidal di setiap stasiun pada saat pengukuran

output dari perhitungan koreksi tidal ini ada 3, yaitu:
1.  **gm** : efek gravitasi dari bulan
2.  **gs** : efek gravitasi dari matahari
3.  **g0** : efek gravitasi total

```python
# menghitung nilai koreksi tidal untuk setiap stasiun pada saat waktu pengukuran
df_tide_sta = solve_tide_df(df_lonlat)

# menampilkan 5 baris pertama dalam tabel
df_tide_sta.head()
```

## 3.3 Menghitung variasi koreksi tidal di base dalam rentang waktu 1 hari per menit

```python
# static data (base)
lat = df_lonlat['lat'][0]
lon = df_lonlat['lon'][0]
alt = z_acuan

# waktu awal
t0 = datetime(2019, 7, 11, 0, 0, 0)

# menghitung koreksi tidal
df_tide_base = solve_point_corr(lat, lon, alt, t0, n=60*12, increment='min')
df_tide_base.head()
```

## 3.4 Visualisasi koreksi tidal

```python
# Plot koreksi tidal
plt.figure(figsize=(12,10))
plt.scatter(df_tide_sta.index, df_tide_sta['g0'], c='k', label='sta')
plt.plot(df_tide_base.index, df_tide_base['g0'], label='base')
plt.xlabel('Time UTC+0', fontsize=18)
plt.ylabel('Tidal Correction [mGals]', fontsize=18)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.xlim([min(df_tide_base.index) - timedelta(minutes=10), max(df_tide_base.index) + timedelta(minutes=10)])
plt.title(f"kelompok {kelompok}")
plt.legend(fontsize=15)
plt.show()
```

# 4. SAVE TABEL KOREKSI ELEVASI DAN KOREKSI TIDAL

```python
# merubah indeks tabel menjadi stasiun untuk digabungkan dengan tabel koreksi elevasi
df_tide = df_tide_sta
df_tide.index = df_elevasi.index

# menggabungkan tabel koreksi elevasi dan tidal
df_koreksi = pd.concat([df_elevasi, df_tide], axis=1)

# menampilkan tabel keseluruhan tabel koreksi
df_koreksi
```
